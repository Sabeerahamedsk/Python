###**Question**: Take an input from the user for start and end values. Print all even numbers between them.

- Start at the number and print if the number is divisible by 2, increment number by 1

````python
start = int(input())
end = int(input())
loop_count = 0
while start <= end:
    loop_count += 1
    if start % 2 == 0:
        print(start)
    start += 1
print("Loop was run", loop_count,"times!")
````

- If number is even, start from number, if odd, start from number+1, print number, increment by 2

````python
start = int(input())
end = int(input())
if start % 2 == 0:
    i = start
else:
    i = start +1
loop_count = 0
while i <= end:
    loop_count += 1
    print(i)
    i += 2
print("Loop was run", loop_count,"times!")
````

- Second answer is more optimized, since the loop runs less number of times.

## Scope of Variables

````python
a = 10 # Global Variable
def random():
    a = 50 # Local Variable
    print("Inside -", end = " ")
    print(a)
random()

print("Outside -", end = " ")
print(a)
````

* The variable ```a``` in the function above is defined within the block of that function. The variable "a" inside the function random is a ```local variable``` and is only accessible inside the function.
* The variable ```a``` outside the function body is accesible anywhere in the code below and hence it is called a ```global variable```.

````python
a = 10 # Global Variable

# A local variable with name "a" is defined inside Random1.
def random1():
    a = 50 # Local Variable
    print("Inside Random1 -", end = " ")
    print(a)
random()

# No local variable with name "a" is defined inside Random2.
def random2():
    print("Inside Random2 -", end = " ")
    print(a)
random2()

print("Outside -", end = " ")
print(a)
````

* If there is **no** local definition of a variable **within** the function, and that variable is called/used from within a function's body, the function would just use the **global variable**.

The `global` keyword in front of a variable name is used to tell python to use the global value of that variable.

````python
a = 10
def random():
    global a
    a = 20
    print(a)
random()
print(a)
````

```runs = [62, 85, 74, 10, 12, 101, 122, 99, 81, 55]```

### **Mutability and Immutability**

`Mutable`: whose value can be changed

`Immutable`: whose value can't be changed

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/016/227/original/Screenshot_2022-10-12_at_10.10.40_AM.png?1665549635">

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/016/228/original/Screenshot_2022-10-12_at_10.10.55_AM.png?1665549675">

````python
a = 3
a
````

````python
a = 4
a
````

It seems that integers are mutable, but NO.

Originally in Python, we are just changing the memory reference of `a` to point to `4` instead of changing its value to `4`.

Only 3 data types in Python are `mutable`.
- `lists`
- `sets`
- `dictionaries`

Other data types such as `strings` and `tuples` are `immutable`.

`id()` function in Python gives the memory address of a variable where it is stored.

````python
id(a)
````

````python
a = 6

id(a)
````

It can be seen that the memory address is changed. Hence, now `a` is pointing to somewhere else in the memory.

````python
my_list =["Python", "C++", "Java"]
id(my_list)
````

````python
my_list[0] = "Python 3.5"
print(my_list)
print(id(my_list))
````

As lists are mutable, we can see that the same list created in the first cell is modified and the values are changed in the same memory location.

````python
my_string  = "Scaled"
print(id(my_string))
````

````python
my_string[-1] = "a"  # Trying to change the character at the last index
````

- We receive an error since strings are immutable and we cannot change the value in place.
- We can instead assign a new value to the same variable, which leads to the creation of a different memory space.

````python
my_string = "Scaler"
print(id(my_string))
````

### **Question - 3**

* You are given a string as input.
* Ascertain whether that string contains a binary number or not.

```
Sample input 1
"0011010101ab"
```

```
Sample output 1
"Not Binary!"
```

```
Sample input 2
"010010101010"
```

```
Sample output 2
"Binary!"
```

````python
# Method 1
def check_binary(string):
    for i in string:
        if not (i == "0" or i == "1"):
            return "Not Binary!"
    return "Binary!"
````

````python
print(check_binary("adasfafasfas"))
print(check_binary("01010101010"))
````

````python
# Method 2
def check_binary_cool(string):
    s = set(string)
    if s == {"0", "1"}:
        return "Binary!"
    return "Not Binary!"
````

````python
print(check_binary_cool("adasfafasfas"))
print(check_binary_cool("01010101010"))
````
